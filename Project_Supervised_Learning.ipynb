We create a knn_scores function that computes and returns performance scores (accuracy, precision, recall, and F1 score) for k-Nearest Neighbors (KNN) classification across a specified range of values for the number of neighbors (k).

It uses a KNeighborsClassifier with user-defined parameters, including the choice of distance metric (default is Minkowski), weighting strategy (default is uniform), and algorithm for tree-based or brute-force search (default is 'auto').

The function fits the KNN model to a training dataset (X_train, Y_train), then predicts class labels on a test dataset (X_test).

For each value of k in the specified range, it calculates accuracy, precision, recall, and F1 score, aggregating the scores into separate DataFrames.

````python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier

def knn_scores(k_range, weights='uniform', algorithm='auto', leaf_size=30, metric='minkowski', p=2):
    accuracy_scores_test = []
    precision_scores_test = []
    recall_scores_test = []
    f1_scores_test = []

    for k in k_range:
        # Create a KNN classifier with the specified parameters
        NNH = KNeighborsClassifier(n_neighbors=k, weights=weights, metric=metric, p=p)
        NNH.fit(X_train, Y_train)

        # Calculate accuracy for the current k on the test data
        Y_pred_test = NNH.predict(X_test)
        accuracy_test = accuracy_score(Y_test, Y_pred_test)

        # Calculate precision for the current k on the test data
        precision_test = precision_score(Y_test, Y_pred_test, average='weighted')

        # Calculate recall for the current k on the test data
        recall_test = recall_score(Y_test, Y_pred_test, average='weighted')

        # Calculate F1 score for the current k on the test data
        f1_test = f1_score(Y_test, Y_pred_test, average='weighted')

        # Append the scores to the respective lists for test data

        accuracy_scores_test.append(accuracy_test)
        precision_scores_test.append(precision_test)
        recall_scores_test.append(recall_test)
        f1_scores_test.append(f1_test)

    df_accuracy_test = pd.DataFrame({'k': k_range, 'accuracy_scores_test': accuracy_scores_test})
    df_precision_test = pd.DataFrame({'k': k_range, 'precision_scores_test': precision_scores_test})
    df_recall_test = pd.DataFrame({'k': k_range, 'recall_scores_test': recall_scores_test})
    df_f1_test = pd.DataFrame({'k': k_range, 'f1_scores_test': f1_scores_test})

    return df_accuracy_test, df_precision_test, df_recall_test, df_f1_test
````

Next we conduct a evaluation of a k-Nearest Neighbors (KNN) classifier across a range of values for the number of neighbors (k), from 1 to 248, usingother metrics as default. Four essential classification performance metrics, including accuracy, precision, recall, and F1 score, are calculated for each k value. Next we plot the metrics vs k to find optimum k value.


````python
k_range = range(1, 248)
df_accuracy_test, df_precision_test, df_recall_test, df_f1_test = knn_scores(k_range)
````

````python
# Define the range of k values (x-axis)
k_values = df_accuracy_test['k']

# Plot
plt.figure(figsize=(12, 8))
plt.plot(k_values, df_accuracy_test['accuracy_scores_test'], label='Accuracy (Test)')
plt.plot(k_values, df_recall_test['recall_scores_test'], label='Recall (Test)', color='g')
plt.plot(k_values, df_precision_test['precision_scores_test'], label='Precision (Test)', color='r')
plt.plot(k_values, df_f1_test['f1_scores_test'], label='F1 Score (Test)', color='y')

plt.title('k vs Metrics (Test Data)')
plt.xlabel('k (Number of Neighbors)')
plt.ylabel('Scores')
plt.legend()  # Show a legend to differentiate the lines
plt.grid(True)
plt.show()
````

We create a function **compute_knn_scores_for_metrics** to perform k-Nearest Neighbors (KNN) classification for various combinations of distance metrics, weights, and algorithms. It calculates classification performance metrics for different scenarios and compiles the results into a pandas DataFrame for analysis.

**Input Parameters:**
*  **k_range**: A range of k values representing the number of neighbors to consider.
*  **metrics:** A list of distance metrics to evaluate, including Euclidean, Manhattan, Chebyshev, Minkowski, Canberra, Bray-Curtis, Jaccard, and more.
*  **weights:** A string specifying the type of weightings to use in KNN, which can be 'uniform' or 'distance'.
*  **algorithms:** A list of algorithms to use, which can be 'auto', 'brute', 'kd_tree', or 'ball_tree'.
*  **p:** An optional parameter specifying the power parameter for the Minkowski metric (default is None).

**Output:**
A pandas DataFrame containing the following columns:
*  **'Metric':** The distance metric being used, e.g., Euclidean, Manhattan, etc.
*  **'P':** The power parameter for the Minkowski metric (only if Minkowski is chosen).
*  **'Algorithm':** The KNN algorithm being used, e.g., 'auto', 'brute', etc.
'K': The number of neighbors (k value).
*  **'Accuracy':** The accuracy score for the KNN classification.
*  **'Precision':** The precision score for the KNN classification.
*  **'Recall':** The recall score for the KNN classification.
*  **'F1':** The F1 score for the KNN classification.

````python
import pandas as pd

def compute_knn_scores_for_metrics(k_range, metrics, weights, algorithms, p=None):
    results = []

    for metric in metrics:
        for algorithm in algorithms:
            if metric == 'minkowski':
                for p_value in range(1, 21):
                    df_accuracy_test, df_precision_test, df_recall_test, df_f1_test = knn_scores(k_range, weights=weights, metric=metric, algorithm=algorithm, p=p_value)
                    for k, accuracy_test, precision_test, recall_test, f1_test, in zip(k_range, df_accuracy_test['accuracy_scores_test'],  df_precision_test['precision_scores_test'],  df_recall_test['recall_scores_test'], df_f1_test['f1_scores_test']):
                        results.append((metric, p_value, k, algorithm, accuracy_test, precision_test, recall_test, f1_test))
            else:
                df_accuracy_test, df_precision_test, df_recall_test, df_f1_test = knn_scores(k_range, weights=weights, metric=metric, algorithm=algorithm)
                for k, accuracy_test, precision_test, recall_test, f1_test, in zip(k_range, df_accuracy_test['accuracy_scores_test'], df_precision_test['precision_scores_test'], df_recall_test['recall_scores_test'], df_f1_test['f1_scores_test']):
                    results.append((metric, None, k, algorithm, accuracy_test, precision_test, recall_test, f1_test))

    df = pd.DataFrame(results, columns=['Metric', 'P', 'K', 'Algorithm', 'Accuracy_Test', 'Precision_Test', 'Recall_Test', 'F1_Test'])

    return df

metrics_to_test = ['euclidean', 'manhattan', 'chebyshev', 'minkowski', 'canberra', 'braycurtis', 'jaccard', 'matching', 'rogerstanimoto', 'dice', 'russellrao', 'sokalsneath', 'cityblock', 'cosine', 'correlation', 'sqeuclidean']
algorithms_to_test = ['auto', 'brute', 'kd_tree', 'ball_tree']

k_range = range(1, 51)  # You can adjust the range as needed

# Compute for weights='distance'
df_distance = compute_knn_scores_for_metrics(k_range, metrics_to_test, weights='distance', algorithms=algorithms_to_test)

# Compute for weights='uniform'
df_uniform = compute_knn_scores_for_metrics(k_range, metrics_to_test, weights='uniform', algorithms=algorithms_to_test)
````

##### **KNN Classifier**

We create a function that takes in various parameters of KNN as input, trains the model for various k-values and returns performance metrics as an output.

````python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier

def knn_scores(k_range, weights='uniform', algorithm='auto', leaf_size=30, metric='minkowski', p=2):
    accuracy_scores_test = []
    precision_scores_test = []
    recall_scores_test = []
    f1_scores_test = []

    for k in k_range:
        # Create a KNN classifier with the specified parameters
        NNH = KNeighborsClassifier(n_neighbors=k, weights=weights, metric=metric, p=p)
        NNH.fit(X_resampled, Y_resampled)

        # Calculate accuracy for the current k on the test data
        Y_pred_test = NNH.predict(X_test)
        accuracy_test = accuracy_score(Y_test, Y_pred_test)

        # Calculate precision for the current k on the test data
        precision_test = precision_score(Y_test, Y_pred_test)

        # Calculate recall for the current k on the test data
        recall_test = recall_score(Y_test, Y_pred_test)

        # Calculate F1 score for the current k on the test data
        f1_test = f1_score(Y_test, Y_pred_test)

        # Append the scores to the respective lists for test data

        accuracy_scores_test.append(accuracy_test)
        precision_scores_test.append(precision_test)
        recall_scores_test.append(recall_test)
        f1_scores_test.append(f1_test)

    df_accuracy_test = pd.DataFrame({'k': k_range, 'accuracy_scores_test': accuracy_scores_test})
    df_precision_test = pd.DataFrame({'k': k_range, 'precision_scores_test': precision_scores_test})
    df_recall_test = pd.DataFrame({'k': k_range, 'recall_scores_test': recall_scores_test})
    df_f1_test = pd.DataFrame({'k': k_range, 'f1_scores_test': f1_scores_test})

    return df_accuracy_test, df_precision_test, df_recall_test, df_f1_test
````

In the code below, we are testing various parameters for the K-Nearest Neighbors (KNN) model, specifically focusing on the following parameters:

*  **k_range:** A range of values for the number of neighbors (k) to test, which affects how many nearest neighbors are considered in the KNN algorithm.

*  **metrics_to_test:** A list of distance metrics to evaluate, such as **'euclidean', 'manhattan', 'chebyshev', 'minkowski'**, and others. These metrics determine how the distance between data points is calculated.

*  **weights:** Weights can be 'uniform' or 'distance', and they control the influence of neighboring points on the classification or prediction.

The output of your code is a **Pandas DataFrame (df_distance and df_uniform)** containing the results of the KNN model's performance for the various combinations of the parameters mentioned above. The DataFrame includes columns for:

*  **Metric:** The distance metric used (e.g., 'euclidean', 'manhattan', etc.).

*  **P:** The value of the 'p' parameter (used with 'minkowski' metric).

*  **K:** The number of neighbors (k).

*  **Accuracy_Test:** The accuracy of the KNN model on the test data.

*  **Precision_Test:** The precision of the KNN model on the test data.

*  **Recall_Test:** The recall of the KNN model on the test data.

*  **F1_Test:** The F1 score of the KNN model on the test data.

This DataFrame allows you to compare and analyze the performance of the KNN model under different configurations, helping you identify the best combination of parameters for your specific problem.

````python
import pandas as pd

def compute_knn_scores_for_metrics(k_range, metrics, weights, p=None):
    results = []

    for metric in metrics:
            if metric == 'minkowski':
                for p_value in range(1, 11):
                    df_accuracy_test, df_precision_test, df_recall_test, df_f1_test = knn_scores(k_range, weights=weights, metric=metric, p=p_value)
                    for k, accuracy_test, precision_test, recall_test, f1_test, in zip(k_range, df_accuracy_test['accuracy_scores_test'],  df_precision_test['precision_scores_test'],  df_recall_test['recall_scores_test'], df_f1_test['f1_scores_test']):
                        results.append((metric, p_value, k, accuracy_test, precision_test, recall_test, f1_test))
            else:
                df_accuracy_test, df_precision_test, df_recall_test, df_f1_test = knn_scores(k_range, weights=weights, metric=metric)
                for k, accuracy_test, precision_test, recall_test, f1_test, in zip(k_range, df_accuracy_test['accuracy_scores_test'], df_precision_test['precision_scores_test'], df_recall_test['recall_scores_test'], df_f1_test['f1_scores_test']):
                    results.append((metric, None, k, accuracy_test, precision_test, recall_test, f1_test))

    df = pd.DataFrame(results, columns=['Metric', 'P', 'K', 'Accuracy_Test', 'Precision_Test', 'Recall_Test', 'F1_Test'])

    return df

metrics_to_test = ['euclidean', 'manhattan', 'chebyshev', 'minkowski', 'canberra', 'braycurtis', 'jaccard', 'matching', 'rogerstanimoto', 'dice', 'russellrao', 'sokalsneath', 'cityblock', 'cosine', 'correlation', 'sqeuclidean']

k_range = range(1, 150)  # You can adjust the range as needed

# Compute for weights='distance'
df_distance = compute_knn_scores_for_metrics(k_range, metrics_to_test, weights='distance')

# Compute for weights='uniform'
df_uniform = compute_knn_scores_for_metrics(k_range, metrics_to_test, weights='uniform')
````

##### **SVM MODEL FINE TUNING**

FIrst we define a function that takes in different parameters of SVM as input, trains the model and returns the performance metrics of both the train and the test data.

````python
X_resampled=X_resampled_main
X_test=X_test_main
````

````python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import pandas as pd
from sklearn.svm import SVC
import warnings
import pandas as pd
from sklearn.exceptions import UndefinedMetricWarning

# Define filters to ignore UndefinedMetricWarning and FutureWarning
warnings.filterwarnings("ignore", category=UndefinedMetricWarning)
warnings.filterwarnings("ignore", category=FutureWarning)

def svm_scores(C=1.0, kernel='rbf', gamma='scale', degree=3, coef0=0.0, random_state=42):

    # Create an SVM classifier with specified hyperparameters
    svm_model = SVC(C=C, kernel=kernel, gamma=gamma, degree=degree, coef0=coef0)

    # Train the model
    svm_model.fit(X_resampled, Y_resampled)

    # Predict on the test data
    Y_pred_test = svm_model.predict(X_test)

    # Predict on the train data
    Y_pred_train = svm_model.predict(X_resampled)

    # Calculate accuracy, precision, recall, and F1 score for test data
    accuracy_test = accuracy_score(Y_test, Y_pred_test)
    precision_test = precision_score(Y_test, Y_pred_test)
    recall_test = recall_score(Y_test, Y_pred_test)
    f1_test = f1_score(Y_test, Y_pred_test)

    # Calculate accuracy, precision, recall, and F1 score for train data
    accuracy_train = accuracy_score(Y_resampled, Y_pred_train)
    precision_train = precision_score(Y_resampled, Y_pred_train)
    recall_train = recall_score(Y_resampled, Y_pred_train)
    f1_train = f1_score(Y_resampled, Y_pred_train)

    return accuracy_test, precision_test, recall_test, f1_test,accuracy_train, precision_train, recall_train, f1_train
````

First, lets check the performance of performance of SVM using **kernel as 'rbf'** for different values of **C and gamma**.

````python
import pandas as pd

# Create an empty DataFrame
results_df_rbf = pd.DataFrame(columns=['C', 'kernel', 'gamma', 'Accuracy_test', 'Precision_test', 'Recall_test', 'F1 Score_test', 'Accuracy_train', 'Precision_train', 'Recall_train', 'F1 Score_train', 'Accuracy_Percentage_Difference', 'Precision_Percentage_Difference', 'Recall_Percentage_Difference', 'F1_Percentage_Difference'])
````

````python
# Define the values of C and gamma to evaluate
C_values = [1, 10, 100, 1000]
gamma_values = [0.1, 0.001, 0.01, 10]

for C in C_values:
    for gamma in gamma_values:
        accuracy_test, precision_test, recall_test, f1_test, accuracy_train, precision_train, recall_train, f1_train = svm_scores(C=C, gamma=gamma, kernel='rbf')

        # Calculate percentage differences
        accuracy_percentage_difference = ((accuracy_test - accuracy_train) / accuracy_train) * 100
        precision_percentage_difference = ((precision_test - precision_train) / precision_train) * 100
        recall_percentage_difference = ((recall_test - recall_train) / recall_train) * 100
        f1_percentage_difference = ((f1_test - f1_train) / f1_train) * 100

        results_df_rbf = results_df_rbf.append({'C': C, 'kernel': 'rbf', 'gamma': gamma, 'Accuracy_test': accuracy_test, 'Precision_test': precision_test, 'Recall_test': recall_test, 'F1 Score_test': f1_test, 'Accuracy_train': accuracy_train, 'Precision_train': precision_train, 'Recall_train': recall_train, 'F1 Score_train': f1_train, 'Accuracy_Percentage_Difference': accuracy_percentage_difference, 'Precision_Percentage_Difference': precision_percentage_difference, 'Recall_Percentage_Difference': recall_percentage_difference, 'F1_Percentage_Difference': f1_percentage_difference}, ignore_index=True)
````
